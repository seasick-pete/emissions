```python
import pandas as pd
uri = '../output_csv_route_specific_2012.csv'

def make_dummy_data(df, yr, factor):
    if factor != 1:
        factor = 1 + (factor /10) 
    print(factor)
    df_y = df.copy()
    df_y.year = yr
    df_y.mean_sog = df_y.mean_sog * factor
    df_y.total_co2_million_tonnes = df_y.total_co2_million_tonnes * factor
    return df_y


df = pd.read_csv(uri, sep='\t')

df_out = pd.concat(
        make_dummy_data(df, yr, idx+1)
        for idx, yr
        in enumerate([2012, 2013, 2014, 2015, 2016]))

df_out.to_csv('../output_csv_route_specific_dummy.csv', '\t')
```
